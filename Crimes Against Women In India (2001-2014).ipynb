### Now we clean the data set

Start by dealing with inconsistent names of States and Union Territories

```python
df['STATE/UT']=df['STATE/UT'].str.lower()
```

```python
df['STATE/UT'].unique()
```

```python
df["STATE/UT"].replace({"delhi ut": "delhi", "a&n islands": "a & n islands", "d&n haveli" :"d & n haveli"}, inplace=True)
```

```python
df['STATE/UT'].unique()
```

```python
df['STATE/UT'].nunique()
```

Ladakh is missing and daman and diu and d and n haveli were combined in 2019

```python
total_df = df[df.DISTRICT == 'TOTAL']
```

```python
yeartotal_df=total_df.groupby('STATE/UT')[['Rape','Kidnapping and Abduction','Dowry Deaths','Assault on women with intent to outrage her modesty','Insult to modesty of Women','Cruelty by Husband or his Relatives','Importation of Girls']].sum()
```

```python
df=df[df["DISTRICT"].str.contains("TOTAL")==False]
```

```python
year_df=df.groupby('Year')[['Rape','Kidnapping and Abduction','Dowry Deaths','Assault on women with intent to outrage her modesty','Insult to modesty of Women','Cruelty by Husband or his Relatives','Importation of Girls']].sum()
```

```python
year_df
```

```python
states_df=df.groupby('STATE/UT')[['Rape','Kidnapping and Abduction','Dowry Deaths','Assault on women with intent to outrage her modesty','Insult to modesty of Women','Cruelty by Husband or his Relatives','Importation of Girls']].sum()
```

```python
states_df['STATE/UT'] = states_df.index
year_df['Year']= year_df.index
```

# Data Analysis:

```python
arr = year_df.sum(axis=1)
sns.set(rc={'figure.figsize':(15,11)})
ax=plt.axes()

plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')

plt.xlabel("Year",fontweight='bold')
plt.ylabel("Total Crimes",fontweight='bold')
plt.title("Distribution of the total number of crimes over the years",fontweight='bold',fontsize=20)

ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)

bar= sns.barplot(x="Year", y=arr, palette="magma_r", data=year_df)
for p in bar.patches:
    bar.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
```

```python
crimes=['Rape','Kidnapping and Abduction','Dowry Deaths',
        'Assault on women with intent to outrage her modesty',
        'Insult to modesty of Women','Cruelty by Husband or his Relatives',
        'Importation of Girls']

df_top_crimes=pd.DataFrame(columns=['crimes',"total"])

for i in crimes:
    df_top_crimes=df_top_crimes.append({'crimes':i ,'total':df[i].sum(axis=0)},ignore_index=True)
    
df_top_crimes=df_top_crimes.sort_values(by='total',ascending=False)

df_top_crimes

list_crimes_number=list(df_top_crimes['total'])
list_crimes=list(df_top_crimes['crimes'])
fig=plt.figure(figsize=(12,10))

ax=plt.axes()
colors = sns.color_palette('coolwarm')
plt.pie(list_crimes_number,labels=list_crimes,autopct='%1.2f%%', colors=colors)

plt.title("Distribution of Crimes from 2001 to 2014",fontweight='bold',fontsize=16)
```

```python
f,ax=plt.subplots(1,2,figsize=(25,10))
colors = sns.color_palette('vlag_r')

ax[0].pie(year_df.iloc[0],autopct='%1.2f%%', colors=colors, )
ax[0].text(0,1.4,"2001",fontweight='bold',fontsize=20)
ax[1].pie(year_df.iloc[12],autopct='%1.2f%%', colors=colors)
ax[0].text(3.4,1.4,"2014",fontweight='bold',fontsize=20)
plt.legend(labels=crimes,loc='best')
```

```python
sns.set(rc={'figure.figsize':(15,11)})
ax=plt.axes()

plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')

plt.xlabel("Year",fontweight='bold')
plt.ylabel("Number of Rapes",fontweight='bold')
plt.title("Total number Of Rapes in each year",fontweight='bold',fontsize=20)

ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)

bar= sns.barplot(x="Year", y="Rape", palette="crest", data=year_df)
for p in bar.patches:
    bar.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

```

```python
df_sorted= states_df.sort_values('Rape')
sns.set(rc={'figure.figsize':(15,11)})
ax=plt.axes()
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)

sns.barplot(x="STATE/UT", y="Rape", palette="crest", data=df_sorted)


plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')
plt.xlabel("States",fontweight='bold')
plt.ylabel("Number of Rapes",fontweight='bold')
plt.title("Statewise Distribution of Rapes in India",fontweight='bold',fontsize=20)
```
