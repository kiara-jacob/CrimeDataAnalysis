# Crimes against Women in India (2001-2014)

From even before the time of her birth a girl could become a victim of a crime. 
The 2012 National Crime Records Bureau report of India states a reported crime rate of 46 per 100,000, rape rate of 2 per 100,000, dowry homicide rate of 0.7 per 100,000 and the rate of domestic cruelty by husband or his relatives as 5.9 per 100,000. 

According to the National Crime Records Bureau, one woman is raped every 20 minutes in India. Incidents of reported rape increased 3% from 2011 to 2012. Incidents of reported incest rape increased 46.8% from 268 cases in 2011 to 392 cases in 2012.

This data set only covers a small subset of all the crimes comitted against women on a daily basis in the country. The number of cases unreported also means this data is not complete. 

The aim of this project is to analyse data, collected between 2001-2014, for any descriptive trends in the crimes comitted against women in India.


In this project, we will focus on finding possible answers to the following questions:

- Are there any trends in the total number of reported cases from 2001 to 2014?
- Which states have the highest and lowest number of cases?
- Based on this data set, which crime is reported the most?
- Are there any trends in crimes for states with the highest cases as compared to those with the lowest?
- Are there in any trends in the most reported crimes?

## 1. Prepping the Data

In order to clean the data set, we first modify the names of the states to be unifor in captilization.

```python
df['STATE/UT']=df['STATE/UT'].str.lower()
```

```python
df["STATE/UT"].replace({"delhi ut": "delhi", "a&n islands": "a & n islands", "d&n haveli" :"d & n haveli"}, inplace=True)
df['STATE/UT'].unique()
```

Ladakh is missing from this data set and daman and diu and d & n haveli were combined in 2019 but have been listed seperately here.

```python
df['STATE/UT'].nunique()
```

Next we need to extract the TOTAL values from the data set since including it will result in double counting. We store the TOTAL values in a new data frame since it is still useful data.

```python
total_df = df[df.DISTRICT == 'TOTAL']
total_df=total_df.drop("DISTRICT", axis=1)
total_df
```

```python
total_df.sample(4).reset_index().drop("index", axis=1)
```

```python
year_df=df.groupby('Year')[['Rape','Kidnapping and Abduction','Dowry Deaths','Assault on women with intent to outrage her modesty','Insult to modesty of Women','Cruelty by Husband or his Relatives','Importation of Girls']].sum()
year_df
```

```python
states_df=df.groupby('STATE/UT')[['Rape','Kidnapping and Abduction','Dowry Deaths','Assault on women with intent to outrage her modesty','Insult to modesty of Women','Cruelty by Husband or his Relatives','Importation of Girls']].sum()
states_df
```

## 2. Identifying and Visualizing Trends

### I. Using Matplotlib and Seaborn to generate graphical representations of the overall data set by year and state.

```python
arr = year_df.sum(axis=1)
sns.set(rc={'figure.figsize':(12,9)})
ax=plt.axes()

plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')

plt.xlabel("Year",fontweight='bold')
plt.ylabel("Total Crimes",fontweight='bold')
plt.title("Fig 1.1 Distribution of the total number of crimes over the years",fontweight='bold',fontsize=20)

ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)

bar= sns.barplot(x="Year", y=arr, palette="coolwarm", data=year_df)
for p in bar.patches:
    bar.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
```

```python
states_df["total_crime"]= states_df.sum(axis=1)
#states_df.drop("STATE/UT", axis=1)
new=states_df.sort_values("total_crime", ascending=False)
sns.set(rc={'figure.figsize':(12,9)})
ax=plt.axes()
new=new.reset_index()

ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
sns.barplot(x="STATE/UT", y="total_crime", palette="coolwarm_r", data=new)
plt.xlabel("States and Union Territories", fontweight='bold')
plt.ylabel("Total Crimes",fontweight='bold')
plt.title(" Fig 1.2 Statewise Distribution of Crime from 2001-2014",fontweight='bold',fontsize=20)
```

```python
crimes=['Rape','Kidnapping and Abduction','Dowry Deaths',
        'Assault on women with intent to outrage her modesty',
        'Insult to modesty of Women','Cruelty by Husband or his Relatives',
        'Importation of Girls']

df_top_crimes=pd.DataFrame(columns=['crimes',"total"])

for i in crimes:
    df_top_crimes=df_top_crimes.append({'crimes':i ,'total':df[i].sum(axis=0)},ignore_index=True)
    
df_top_crimes=df_top_crimes.sort_values(by='total',ascending=False)


list_crimes_number=list(df_top_crimes['total'])
list_crimes=list(df_top_crimes['crimes'])
fig=plt.figure(figsize=(11,9))

ax=plt.axes()
colors = sns.color_palette('coolwarm',6)
plt.pie(list_crimes_number,labels=list_crimes,autopct='%1.2f%%', colors=colors,counterclock=False, center=(1,0))

plt.title("Fig 1.3 Distribution of Crimes from 2001 to 2014",fontweight='bold',fontsize=20)
```

```python
f,(ax1,ax2)=plt.subplots(1,2)
f.set_figheight(12)
f.set_figwidth(12)

colors = sns.color_palette('coolwarm',8)
ax1.pie(year_df.iloc[0],autopct='%1.2f%%', colors=colors,radius=1.5, pctdistance=0.80, center=(1.5,0),counterclock=True)
ax1.text(0.1,1.7,"Fig 1.4 Crime Distrubution in 2001",fontweight='bold',fontsize=20)
ax2.pie(year_df.iloc[13],autopct='%1.2f%%', colors=colors,radius=1.5, pctdistance=0.80, center=(1.5,0),counterclock=True)
plt.text(0.1,1.7,"Fig 1.5 Crime Distrubution in 2014",fontweight='bold',fontsize=20)
plt.legend(loc="center", bbox_to_anchor=(0.5, -0.25), ncol= 2, labels=crimes)
f.tight_layout()
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=1, 
                    top=1, 
                    wspace=0.5, 
                    hspace=0.5)
```

```python
sns.set(rc={'figure.figsize':(17,10)})
colors = sns.color_palette('hls')
line_df=year_df
line_df=line_df.drop("Year", axis=1)
line_df.plot(color=colors, xlabel="YEAR", ylabel="Crime Rate", kind="line", fontsize=10,linewidth=3);
plt.title("Fig 1.6 Distribution of various crimes over the years", fontsize=20,fontweight='bold')
```

```python
transposed_df=year_df
transposed_df=transposed_df.transpose()
transposed_df["total_crime"]= transposed_df.sum(axis=1)
#transposed_df.drop("total_crime", axis=0)

temp_df=transposed_df["total_crime"]
colors = sns.color_palette('flare')

plt.ylabel("Total Crimes",fontweight='bold')
temp_df=temp_df.sort_values(ascending=False).drop("Year")
temp_df.plot(kind="bar", color=colors)
plt.title("Fig 1.7 Distribution of total number of cases by crime",fontweight="bold", fontsize=20)
```
