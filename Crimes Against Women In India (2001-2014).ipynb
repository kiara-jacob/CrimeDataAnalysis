# Crimes against Women in India (2001-2014)

From even before the time of her birth a girl could become a victim of a crime. 
The 2012 National Crime Records Bureau report of India states a reported crime rate of 46 per 100,000, rape rate of 2 per 100,000, dowry homicide rate of 0.7 per 100,000 and the rate of domestic cruelty by husband or his relatives as 5.9 per 100,000. 
A 2014 study in the Lancet states, “Whereas an 8·5% prevalence of sexual violence in the country [India] is among the lowest in the world, it is estimated to affect 27·5 million women in India [given India’s large population]. Further, the 2006 survey found that 85% of women who suffered sexual violence, in or outside of marriage, never sought help, and only 1% report it to the police.” 
This data set only covers a small subset of all the crimes comitted against women on a daily basis in the country. The number of cases unreported also means this data is not complete. 
The aim is to analyse it for descriptive trends in the reported cases from all over the country between 2001-2014.

SOURCE: http://www.womenlawsindia.com/legal-awareness/crimes-against-women/

### Now we clean the data set

# Data Analysis:

# OVERALL ANALYSIS

```python
arr = year_df.sum(axis=1)
sns.set(rc={'figure.figsize':(12,9)})
ax=plt.axes()

plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')

plt.xlabel("Year",fontweight='bold')
plt.ylabel("Total Crimes",fontweight='bold')
plt.title("Distribution of the total number of crimes over the years",fontweight='bold',fontsize=20)

ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)

bar= sns.barplot(x="Year", y=arr, palette="coolwarm", data=year_df)
for p in bar.patches:
    bar.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
```

```python
states_df["total_crime"]= states_df.sum(axis=1)
#states_df.drop("STATE/UT", axis=1)
new=states_df.sort_values("total_crime")
sns.set(rc={'figure.figsize':(17,10)})
ax=plt.axes()
new=new.reset_index()

plt.title("Statewise Distribution of Crime from 2001-2014",fontweight='bold',fontsize=20)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
plt.ylabel("Total Crimes",fontweight='bold')
sns.barplot(x="STATE/UT", y="total_crime", palette="coolwarm", data=new)
```

```python
crimes=['Rape','Kidnapping and Abduction','Dowry Deaths',
        'Assault on women with intent to outrage her modesty',
        'Insult to modesty of Women','Cruelty by Husband or his Relatives',
        'Importation of Girls']

df_top_crimes=pd.DataFrame(columns=['crimes',"total"])

for i in crimes:
    df_top_crimes=df_top_crimes.append({'crimes':i ,'total':df[i].sum(axis=0)},ignore_index=True)
    
df_top_crimes=df_top_crimes.sort_values(by='total',ascending=False)

df_top_crimes

list_crimes_number=list(df_top_crimes['total'])
list_crimes=list(df_top_crimes['crimes'])
fig=plt.figure(figsize=(12,10))

ax=plt.axes()
colors = sns.color_palette('coolwarm')
plt.pie(list_crimes_number,labels=list_crimes,autopct='%1.2f%%', colors=colors,counterclock=False)

plt.title("Distribution of Crimes from 2001 to 2014",fontweight='bold',fontsize=20)
```

```python
f,(ax1,ax2)=plt.subplots(1,2)
f.set_figheight(12)
f.set_figwidth(12)

colors = sns.color_palette('coolwarm')
ax1.pie(year_df.iloc[0],autopct='%1.2f%%', colors=colors,radius=1.5, pctdistance=0.80, center=(1,0),counterclock=True)
ax1.text(0.1,1.7,"Crime Distrubution in 2001",fontweight='bold',fontsize=20)
ax2.pie(year_df.iloc[13],autopct='%1.2f%%', colors=colors,radius=1.5, pctdistance=0.80, center=(1,0),counterclock=True)
plt.text(0.1,1.7,"Crime Distrubution in 2014",fontweight='bold',fontsize=20)
plt.legend(loc="center", bbox_to_anchor=(0.5, -0.25), ncol= 2, labels=crimes)
f.tight_layout()
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=1, 
                    top=1, 
                    wspace=0.5, 
                    hspace=0.5)
```

```python
sns.set(rc={'figure.figsize':(17,10)})
colors = sns.color_palette('hls')
year_df.plot(color=colors, xlabel="YEAR", ylabel="Crime Rate", kind="line", fontsize=10);
plt.title("Distribution of various crimes over the years", fontsize=20,fontweight='bold')
```

```python
transposed_df=year_df
transposed_df=transposed_df.transpose()
transposed_df["total_crime"]= transposed_df.sum(axis=1)
#transposed_df.drop("total_crime", axis=0)

temp_df=transposed_df["total_crime"]
colors = sns.color_palette('flare')

plt.ylabel("Total Crimes",fontweight='bold')
plt.xlabel("Crime",fontweight='bold')
temp_df=temp_df.sort_values(ascending=False).drop("Year")
temp_df.plot(kind="bar", color=colors)
```
